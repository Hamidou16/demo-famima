```python
df_PCE = df_join.loc[df_join.item_name == 'PC-E'].copy()
# df_PCE.payment_month = pd.PeriodIndex(df_PCE.payment_month, freq='M')
df_PCE
```

```python
PCE_sum = df_PCE.groupby('payment_month').sum()['item_price']
PCE_sum
```

```python
df_PCE_sum = pd.DataFrame(PCE_sum)
df_PCE_sum.reset_index(level=0, inplace=True)
df_PCE_sum
```

```python
df_PCE_sum['payment_month'] = df_PCE_sum.payment_month.astype(float)
```

```python
df_PCE_sum_item_price = df_PCE_sum['item_price']
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df_PCE_sum_item_price)
print('ADF Statistics: %f' %result[0])
print('p-value: %f' %result[1])
```

Since P-value is greater than the significance level ($p = 0.05$), let's difference the series and see how the autocorrelation plot looks like

```python
# original series
fig, ax = plt.subplots(2)
fig.suptitle('Original series')
ax[0].plot(df_PCE_sum_item_price)
plot_acf(df_PCE_sum_item_price, ax=ax[1])
plt.show()
```

```python
# 1st differencing
fig, ax = plt.subplots(2)
fig.suptitle('1st differencing')
ax[0].plot(df_PCE_sum_item_price.diff().values)
plot_acf(df_PCE_sum_item_price.diff()[1:].values, ax=ax[1])
plt.show()
```

```python
# 2nd differencing
fig, ax = plt.subplots(2)
fig.suptitle('2nd differencing')
ax[0].plot(df_PCE_sum_item_price.diff().diff().values)
plot_acf(df_PCE_sum_item_price.diff().diff()[2:].values, ax=ax[1])
plt.show()
```

```python
from pmdarima.arima.utils import ndiffs
y = df_PCE_sum_item_price
# KPSS test
print(ndiffs(y, test='kpss'))
# PP test
print(ndiffs(y, test='pp'))
```

```python
# 1,1,1 ARIMA model
model = ARIMA(df_PCE_sum_item_price, order=(1,1,0))
model_fit = model.fit()
print(model_fit.summary())
```

```python
# 1,2,0 ARIMA model
model = ARIMA(df_PCE_sum_item_price, order=(1,2,0))
model_fit = model.fit()
print(model_fit.summary())
```

```python
# plot residual errors
residuals = pd.DataFrame(model_fit.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```

The residual errors seem fine with near $0$ mean and uniform variance.

```python
# Accuracy metrics
from statsmodels.tsa.stattools import acf
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    corr = np.corrcoef(forecast, actual)[0,1]   # corr
    mins = np.amin(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    maxs = np.amax(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)             # minmax
    acf1 = acf(fc-test)[1]                      # ACF1
    return({'mape':mape, 'me':me, 'mae': mae, 
            'mpe': mpe, 'rmse':rmse, 'acf1':acf1, 
            'corr':corr, 'minmax':minmax})

forecast_accuracy(fc, test.values)
```

Around $12\%$ MAPE implies that the model is about $88\%$ accurate about predicting the next $2$ months observations.
